```python
L = [[5/4, -3/2], [-1.5, 5]]
```

```python
L
```

```python
R = np.random.randn(500,2)
```

```python
for i in range(0,500):
    R[i] = R[i] / np.linalg.norm(R[i])
```

```python
R = R.T
```

```python
R
```

```python
iii = np.matmul(L, R)
```

```python
iii
```

```python
np.linalg.eig(L)
```

```python
v = []
for i in range(0, 500):
    v.append(np.linalg.norm(iii[:,i]))
```

```python
v
```

```python
eign = [0.72382841, 5.52617159]
```

```python
plt.hist(v, bins=50, edgecolor="black", alpha=0.5)
plt.hist(eign, bins=50, color = 'red', edgecolor="black", alpha=0.5)
plt.show()
```

```python
origin = [0], [0] # origin point
plt.quiver(*origin, iii2[:,0], iii2[:,1], scale=18)
plt.quiver(*origin, Lvmax[0], Lvmax[1], color=['r'], scale=18)
plt.show()
```
